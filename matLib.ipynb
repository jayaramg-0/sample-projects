```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [5, 7, 8, 10, 12]
y = [10, 14, 15, 18, 20]

plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
x = ['Apple', 'Banana', 'Cherry']
y = [30, 45, 10]

plt.bar(x, y, color='orange')
plt.title("Fruit Sales")
plt.xlabel("Fruit")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
x = ['Python', 'Java', 'C++']
y = [90, 75, 60]

plt.barh(x, y, color='skyblue')
plt.title("Programming Language Popularity")
plt.xlabel("Score")
plt.ylabel("Language")
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=20, color='green', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
sizes = [40, 35, 25]
labels = ['Python', 'Java', 'C++']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Language Popularity Share")
plt.show()
```

```python
data = [5, 8, 10, 15, 18, 20, 22, 25, 30]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

```python
x = [1, 2, 3, 4]
plt.plot(x, [1, 4, 9, 16], label='Square')
plt.plot(x, [1, 8, 27, 64], label='Cube')

plt.legend()
plt.title("Legend Example")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.annotate("Peak Point", xy=(5, 10), xytext=(3, 9),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```
